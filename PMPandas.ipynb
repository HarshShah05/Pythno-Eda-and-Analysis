DropNa---to drop missing values

```python
# axis =1---columns
# axis=0---rows
import pandas as pd
import numpy as np
dict2={
    "names":["a","b","c","d","e","f","g",np.nan],
    "Marks":[98,np.nan,99,87,90,np.nan,82,np.nan],
    "Grades":[np.nan,"AB","AA",np.nan,"AC","BA","BB",np.nan]
}


df2=pd.DataFrame(dict2)
df2
```

```python
# drop rows atleast having one missing value
# by default axis is 0

df2.dropna(how="any")
```

```python
# drop columns  atleast having one missing value

df2.dropna(how="any",axis=1)
# all columns dropped coz all  have null values
```

```python
# drop rows where all are missing

df2.dropna(how="all",axis=0)
# 7th row dropped
```

```python
# # drop columns where all are missing
# no such columns

df2.dropna(how="all",axis=1)
```

```python
# Keep rows with atleast 2 non-na values
# only last row having 3 na values
# requires thresh parameter

df2.dropna(axis=0,thresh=2)
```

```python
df2.dropna(axis=0,thresh=3)
# rows having 3 non-na values
```

```python
# define in which column to look for missing/na values
# subset we will use

df2.dropna(subset=["Marks"],how="any",axis=0)
```

```python
df2.dropna(subset=["Marks","Grades"],how="any",axis=0)
```

```python
df2.dropna(subset=["names","Grades"],how="any",axis=0)
```

```python
# if (inplace  =true ) then  it modifies original dataframe
# df2.dropna(subset=["names","Grades"],how="any",axis=0,inplace=True)
```
