```python
# which day has highest sales???


# find day of week
# first parse the day using parse_dates

# did before
# df=pd.read_csv(path,parse_dates=["Date"])



dict1={
    0:"Mon",
    1:"Tue",
    2:"Wed",
    3:"Thrus",
    4:"Friday",
    5:"Sat",
    6:"Sunday"
}

print(dict1)
```

```python
df["Date"].dt.dayofweek.map(dict1)
```

```python
df["Day of week"]=df["Date"].dt.dayofweek.map(dict1)
df.head(1)
```

```python
df["Day of week"].value_counts().plot(kind="bar")
# highest sales on saturday
```

```python
explode=[0.3,0.2,0,0,0,0,0]
df["Day of week"].value_counts().plot(kind="pie",autopct="%1.2f%%",explode=explode,shadow=True)
```
