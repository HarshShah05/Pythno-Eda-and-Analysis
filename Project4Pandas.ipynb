Handle missing values

```python
# embarked column has 2 missing values so replace them by mode

df["Embarked"].mode()
```

```python
# fillna used to fill missing values
# mode used for handling categorical value
df["Embarked"].fillna('S',inplace=True)
```

```python
df.isnull().sum()
```

```python
# Handle missing  in numerical values ie Age

df["Age"].mean()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# categorical data encoding
# sex columns and embarked columns has categorical data so encode it

df["Embarked"].value_counts()
```

```python
df.head()
```

```python
# use map for encoding


df["Encoded Embarked"]=df["Embarked"].map({"S":1,"C":2,"Q":3})

df.head(2)
```

```python
def replace1(i):
  if i=="S":
    return 1
  elif i=="C":
    return 2
  else:
    return 3



df["secondencode"]=df["Embarked"].apply(replace1)
```
