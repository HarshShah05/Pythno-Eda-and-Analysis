```python
# Find average app rating

df["Rating"].mean()
```

```python
# Find total number of unique category

len(df["Category"].value_counts())
# df["Category"].nunique()---34
```

```python
# Which category getting highest rating

df[df["Rating"].max()==df["Rating"]]["Category"]
```

```python
# find total  number of apps having 5 star rating

len(df[df["Rating"]==5])
```

```python
df[df["Rating"]==5]
```

```python
# Total number of free and paid apps
df["Type"].value_counts()
```

```python
# Which app has maximum reviews

df[df["Reviews"].max()==df["Reviews"]]
```

```python
# Display top 5 apps having highest reviews

index=df["Reviews"].sort_values(ascending=False).head(5).index
```

```python
df.iloc[index]["App"]
```

```python
# find average rating of free and paid apps
# group by

df.groupby('Type')['Rating'].mean()
```
